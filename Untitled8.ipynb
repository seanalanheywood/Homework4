```python
#Imports all of the necessary modules we will use in this homework assignment and ensures that the plots are generated inline. 
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def homework4():
    data = (pd.read_csv('iris.txt')) #reads in the dataset
    head = data.head() #checks the header to so we have an idea what the data looks like
    print(head)
    reduced_data = data.copy() #copies the data set and stores under reduced data set to be used for the rest of the assignment
    #delete the unwanted columns in the new reduced_data
    del reduced_data['SepalLength']
    del reduced_data['SepalWidth']
    del reduced_data['Species']
    print(reduced_data)
    reduced_data.plot.scatter(x='PetalLength', y='PetalWidth') #creates a scatter plot with x as the PetalLength and y as the PetalWidth
homework4()
```

The header shows that this is usable data.

X = independent variable PetalLength
y = dependent variable PetalWidth
We have a positive relationship between the two variables even with the noticeable gap.
There doesn't appear to be any outliers in this data.

Step 5: Implement the basic functions to create a regression model.
1. Create a function to calculate the coefficients of a least squares approximation.
2. Create a function to use the least squares coefficients to predict the value of a given input point.
Step 6: Using your regression model, calculate the points along the regression line of the reduced data set.
Step 7: Plot the regression line on top of a scatter plot of the reduced data set (you will need to remove duplicate X values to get matplotlib to draw an actual line).

```python
def homework4():
    def beta(PetalLength, PetalWidth):
        beta = np.dot(np.dot(np.linalg.inv(np.dot(np.transpose(PetalLength),PetalLength)),np.transpose(PetalLength)),PetalWidth)
        return beta #np.linalg.inv computes the multiplicative inverse of a matrix.
    ind_var = reduced_data['PetalLength'].values #pulls the x variables
    ind_var_array = np.repeat(1,150) #150 observations in data set
    ind_var_matrix = np.column_stack((ind_var_array, ind_var)) #takes a sequence of 1-D arrays and stack them as columns to make a single 2-D array.
    dep_var = reduced_data['PetalWidth'].values #pulls the y variables
    coefficients = beta(ind_var_matrix,dep_var)
    print(coefficients)
    def least_squares(ind_array, x):
        predicted_var = ind_array[0]+ind_array[1]*x
        return predicted_var
    predicted_array=least_squares(coefficients, 1.1) #1.1 arbitrary point to select for testing
    print(predicted_array)
    y_hat_array = np.empty(shape=[1,2])
    for x in reduced_data['PetalLength']:
        y_hat = least_squares(coefficients, x)
        y_hat_array = np.vstack([y_hat_array, [x, y_hat]])
    y_hat_array = np.delete(y_hat_array,0,0) #This is important so we don't have two lines in the regression in the next step. 
    print(y_hat_array)    
    #Finally, we plot the regression line on the scatter plot.
    y_hat = pd.DataFrame(data=y_hat_array[0:,0:])
    y_hat = y_hat.drop_duplicates()
    plt.scatter(reduced_data['PetalLength'], reduced_data['PetalWidth'])
    plt.plot(y_hat[0],y_hat[1])
    plt.show()
homework4()
```

We have found the coefficients for PetalLength and PetalWidth. They are -.37 and .42 respectively. 
When we use 1.1 as a test for PetalLength, we predict that the PetalWidth would be .09. 
We've also created a new array which stores the x and y values for all observations. This helps us identify the values being plotted on the graph.
We have successfully plotted the regression line on top of the scatter plot.
